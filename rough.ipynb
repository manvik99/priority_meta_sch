```python
import pandas as pd
```

```python
df = pd.read_csv('stored_nice_values.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(10, 6))

# Plot each process's nice value change over time
for pid in df['pid'].unique():
    pid_data = df[df['pid'] == pid]
    plt.plot(pid_data.index, pid_data['nice'], label=f"PID {pid}", marker='o', linestyle='-', markersize=5)

# Customize the plot
plt.title("Change in Nice Values Over Time for Each Process ID", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Nice Value", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Process ID")

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Set up Seaborn style
sns.set(style="whitegrid")

# Plotting with Seaborn
plt.figure(figsize=(8, 4))

# Plot each process's nice value change over time using Seaborn lineplot
sns.lineplot(x=df.index, y=df['nice'], hue=df['pid'], marker='o', palette='tab10', linewidth=2)

# Customize the plot
# plt.title("Change in Nice Values Over Time for Each Process ID", fontsize=16)
plt.xlabel("Time", fontsize=14)
plt.ylabel("Nice Value", fontsize=14)
plt.xticks([])
plt.annotate('', xy=(1, 0), xytext=(0, 0), xycoords='axes fraction', textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.grid(True)
plt.legend(title="Process ID", loc='upper right',fontsize=14, title_fontsize=14)

# Save the plot as an image file
plt.tight_layout()
plt.savefig("nice_values_plot.png")

# Display the plot
plt.show()
```
